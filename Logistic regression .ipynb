```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
    
```

```python
def costFunction(theta, X, y):
    m = y.shape[0]
    theta = theta.reshape(-1, 1)  # Ensure theta is column vector
    y = y.values.reshape(-1, 1) if isinstance(y, pd.Series) else y
    X = X.values if isinstance(X, pd.DataFrame) else X

    h = sigmoid(X @ theta)

    j = (-1/m) * (y.T @ np.log(h) + (1 - y).T @ np.log(1 - h))
    j = j.item()  # Extract scalar

    if np.isnan(j):
        return np.inf
    return j
```

```python
def gradient(theta, X, y):
    m = np.size(y)
    hprime=X.dot (theta)
    h = sigmoid (hprime)
    
    grad = (-1/m)*X.T.dot(h-y)

    return grad.values.flatten()
```

```python
initial_theta = np.zeros((X.shape[1],1))
cost = costFunction(initial_theta, X, y)
grad = gradient(initial_theta, X, y)
print('Cost: \n', cost)
print('Grad: \n', grad)
```

```python
def gradientDescent(X, y, theta=None, alpha=0.01, num_iters=50):
    if theta is None:
        theta = np.zeros((X.shape[1], 1))  # Automatically match number of features

    m = len(y)
    j_history = np.zeros(num_iters)

    for i in range(num_iters):
        h = sigmoid(X.dot(theta))
        gradient = (1 / m) * X.T.dot(h - y)
        theta = theta - alpha * gradient
        j_history[i] = float((-1 / m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h)))

    return theta, j_history
```

```python
theta, j_history = gradientDescent(X, y)
```

```python
def predict_class(user_input, theta):
    if user_input.ndim == 1:
        user_input = user_input.reshape(1, -1)

    prediction = sigmoid(user_input.dot(theta))
    return int((prediction >= 0.5)[0][0])
```
